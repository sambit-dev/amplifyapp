```
import matplotlib.pyplot as plt
import numpy as np
import math
x=np.arange(0,math.pi*2,0.05)
y=np.sin(x)
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin')
plt.plot(x,y)
plt.show()
```

```
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.set_title('plot')
ax.set_xlabel('x')
ax.set_ylabel('y')
x=np.arange(0,math.pi*2,0.05)
y=np.sin(x)
ax.plot(x,y)
plt.show()
```
